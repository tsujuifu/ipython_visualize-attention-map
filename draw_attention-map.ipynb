```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import cv2
```

```python
def softmax(arr):
    out = np.exp(arr)/np.exp(arr).sum()
    return out
```

```python
for i in range(3):
    img = plt.imread('imgs/%d.jpg'%(i))
    
    attn = np.zeros((4, 4))
    for _ in range(2):
        attn[np.random.randint(4)][np.random.randint(4)] = np.random.rand()+0.5
        
    attn = softmax(attn)
    attn = cv2.resize(attn, dsize=(img.shape[1], img.shape[0]), interpolation=cv2.INTER_CUBIC)
    
    plt.figure()
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.show()
    plt.close()
    
    plt.figure()
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.imshow(attn, cmap='gray', alpha=0.9)
    plt.show()
    plt.close()
    
    plt.figure()
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.imshow(attn, alpha=0.9)
    plt.show()
    plt.close()
```
