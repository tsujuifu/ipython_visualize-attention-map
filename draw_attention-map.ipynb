```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import cv2
```

```python
def softmax(arr):
    out = np.exp(arr)/np.exp(arr).sum()
    return out

for i in range(3):
    img = plt.imread('imgs/%d.jpg'%(i))
    
    attn = np.random.rand(64, 64)
    attn = softmax(attn)
    
    attn = cv2.resize(attn, dsize=(img.shape[1], img.shape[0]))
    
    plt.figure()
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.imshow(attn, cmap='gray', alpha=0.4)
    plt.show()
    plt.close()
```
